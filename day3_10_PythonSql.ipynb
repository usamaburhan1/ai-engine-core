**Handling missing data — isnull(),NaN, dropna(), fillna()**

```python
import numpy as np
#create a data frame with missing values
df2 = pd.DataFrame({
    "name": ["Alex", "Maria", "James","Sarah","Tom"],
    "sales":[4200,None,3100,9200,None],
    "city":["Atlanta", "New York", None, "Chicago", "New York"]

})
print(df2)
print("\nMissing values per column:")
print(df2.isnull().sum())
```

```python
#Ways to handle missing data

#Drop rows with any missing value
print(df2.dropna())
print("============")

#fill missing numbers with mean
df2["sales"] = df2["sales"].fillna(df2["sales"].mean())
print(df2)
print("============")

# Option 3 — fill missing text with a default
df2["city"] = df2["city"].fillna("Unknown")
print(df2)
```

The general rule:
Datatype  Strategy

Numbers   Fill with mean, median, or 0 depending on context\n
Text      Fill with "Unknown" or the most common value \n

Critical columnDrop the row entirely

```python
# Multiple stats at once
print(df.groupby("city")["sales"].agg(["mean", "sum", "count"]))

# Group by city, get stats on all numeric columns
print(df.groupby("city").mean(numeric_only=True))
```

Day 3 Final
Fill missing sales with the column mean

Add a column sales_per_month = sales divided by months_active

Filter only rows where sales_per_month is above 400

Group by city — find total sales and average months_active per city

Group by product — find the top performer in each product category

```python
import pandas as pd
import numpy as np

data = {
    "name": ["Alex", "Maria", "James", "Sarah", "Tom", "Lisa"],
    "city": ["Atlanta", "New York", "Atlanta", "Chicago", "New York", "Chicago"],
    "sales": [4200, 7800, None, 9200, 5600, 8100],
    "months_active": [6, 12, 3, 24, 8, 18],
    "product": ["software", "hardware", "software", "hardware", "software", "hardware"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["sales"] = df["sales"].fillna(df["sales"].mean())
df["sales_per_month"] = (df["sales"]/df["months_active"])
print(df)
df2 = df[df["sales_per_month"]>400]
print(df2)
print("++++++++++++++")
print(df.groupby("city")["sales"].agg("sum","mean"))
print(df.groupby("city").agg({"sales":"sum","months_active":"mean"}))
```

```python
print(df.sort_values("sales",ascending=False).groupby("product").first())
```
